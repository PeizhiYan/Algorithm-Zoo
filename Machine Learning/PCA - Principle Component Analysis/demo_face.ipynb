```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import h5py
from pca import PCA
```

```python
# Load Olivetti Faces dataset
images = np.reshape(np.load("./dataset/olivetti_faces.npy"), [-1, 64*64])
print(images.shape)
    
# Display some samples
plt.figure(figsize=(10,2))
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(np.reshape(images[i],[64,64]), cmap='gray')
    plt.axis("off")
```

```python
pca = PCA(method='cov') # use covariance method

x_pca = pca.fit(images, k=100) # fit the PCA model
```

```python
# plot the first 50 transformed testing samples
plt.imshow(x_pca[:50], cmap='gray')
```

```python
# plot the covariance matrix
plt.figure(figsize=(6,6))
plt.imshow(pca.cov, cmap='gray')
```

```python
# Better visualize the covariance matrix (first 20 rows) by reshaping it
plt.figure(figsize=(10,3))
for i in range(3):
    for j in range(10):
        plt.subplot(3,10,i*10+j+1)
        plt.imshow(np.reshape(pca.cov[i*10+j], [64,64]), cmap='gray')
        plt.axis("off")
```

```python
# Show the first 100 principle components (eigenfaces)
plt.figure(figsize=(10,10))
for i in range(10):
    for j in range(10):
        plt.subplot(10,10,i*10+j+1)
        plt.imshow(np.reshape(pca.pc[:,i*10+j],[64,64]), cmap='gray')
        plt.axis("off")
```

```python
# Now we attempt to transform the compressed data back

# Display some samples
plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(4,10,i+1)
    plt.imshow(np.reshape(images[i],[64,64]), cmap='gray')
    plt.axis("off")

# k = 10
test_x_pca = pca.transform(images, k=10) # transform testing data
test_x_inv_pca = pca.inverse_transform(test_x_pca) # transform the compressed data back
for i in range(10):
    plt.subplot(4,10,10+i+1)
    plt.imshow(np.reshape(test_x_inv_pca[i],[64,64]), cmap='gray')
    plt.axis("off")

# k = 50
test_x_pca = pca.transform(images, k=50) # transform testing data
test_x_inv_pca = pca.inverse_transform(test_x_pca) # transform the compressed data back
for i in range(10):
    plt.subplot(4,10,20+i+1)
    plt.imshow(np.reshape(test_x_inv_pca[i],[64,64]), cmap='gray')
    plt.axis("off")
    
# k = 100
test_x_pca = pca.transform(images, k=100) # transform testing data
test_x_inv_pca = pca.inverse_transform(test_x_pca) # transform the compressed data back
for i in range(10):
    plt.subplot(4,10,30+i+1)
    plt.imshow(np.reshape(test_x_inv_pca[i],[64,64]), cmap='gray')
    plt.axis("off")
```
