```python
import torch
torch.manual_seed(666)
import torch.nn as nn
import torch.nn.functional as F
import torch.utils
import torch.distributions
import torchvision
import numpy as np
import matplotlib.pyplot as plt

import tensorflow.compat.v1 as tf           # Tensorflow version 1 from version 2
tf.disable_v2_behavior()

device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
print('Torch device:', device)
```

```python
mnist_dataset = tf.keras.datasets.mnist
(x_train, _), _ = mnist_dataset.load_data() # only load training images
x_train = (x_train / 255.0) # normalize the value to -0 ~ 1.0
x_train = x_train.astype(np.float32)
print('x_train shape:', x_train.shape) # display the shape of training data
print(x_train.dtype)


# display some samples
plt.figure()
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(x_train[i])
    plt.axis('off')
```

```python
class Encoder(nn.Module):
    def __init__(self, latent_dims):
        super(Encoder, self).__init__()
        self.fc1 = nn.Linear(784, 512)
        self.fc2 = nn.Linear(512, 128)
        self.fc3_mu = nn.Linear(128, latent_dims)
        self.fc3_std = nn.Linear(128, latent_dims)
        
    def forward(self, x):
        h = torch.flatten(x, start_dim=1) # flatten input image
        h = F.relu(self.fc1(h))
        h = F.relu(self.fc2(h))
        mu = self.fc3_mu(h)
        sigma = torch.exp(self.fc3_std(h))
        return mu, sigma
    
class Decoder(nn.Module):
    def __init__(self, latent_dims):
        super(Decoder, self).__init__()
        self.fc1 = nn.Linear(latent_dims, 128)
        self.fc2 = nn.Linear(128, 512)
        self.fc3 = nn.Linear(512, 784)
        
    def forward(self, z):
        h = F.relu(self.fc1(z))
        h = F.relu(self.fc2(h))
        x_ = torch.sigmoid(self.fc3(h))
        return x_.reshape(-1, 1, 28, 28)
    
    
class VAE(nn.Module):
    def __init__(self, latent_dims):
        super(VAE, self).__init__()
        self.encoder = Encoder(latent_dims)
        self.decoder = Decoder(latent_dims)
        
        self.normal = torch.distributions.Normal(0, 1) # a normal distribution with mean = 0, std = 1
        if device == 'cuda:0':
            self.normal.loc = self.normal.loc.cuda() # use CUDA GPU for sampling
            self.normal.scale = self.normal.scale.cuda()
        
        self.kl_loss = 0 # KL divergence loss
        
    def forward(self, x):
        mu, sigma = self.encoder(x)
        
        z = mu + sigma * self.normal.sample(mu.size())
        
        self.kl_loss = (sigma**2 + mu**2 - torch.log(sigma) - 0.5).sum()
        
        x_ = self.decoder(z)
        return x_
    
    
```

```python
def compute_reconstruction_loss(x, x_):
    ## squared error
    return ((x.view(-1, 784) - x_.view(-1, 784))**2).sum((1)).mean()
    ##return F.binary_cross_entropy(x_.view(-1, 784), x.view(-1, 784), reduction='sum')
```

```python
vae = VAE(latent_dims=2).to(device)

print('model on CUDA: ', next(vae.parameters()).is_cuda) # True
```

```python
## Train VAE

EPOCHS = 20
LR = 1e-3
BATCH_SIZE = 128

optimizer = torch.optim.Adam(vae.parameters(), lr=LR)

for epoch in range(EPOCHS):
    idx = 0
    while idx < len(x_train):
        
        batch_x = x_train[idx: idx + BATCH_SIZE]
        idx += BATCH_SIZE
        
        x = torch.from_numpy(batch_x).to(device) # create input tensor

        optimizer.zero_grad()
        
        x_ = vae(x)
        
        reconstruction_loss = compute_reconstruction_loss(x, x_) 
        kl_loss = vae.kl_loss
                
        loss = reconstruction_loss + 1e-2*kl_loss
        loss.backward()
                
        optimizer.step()
        
    print(epoch, reconstruction_loss.data, kl_loss.data)
    
    
```

```python
def plot_reconstructed(vae, r0=(-3, 3), r1=(-3, 3), n=30):
    w = 28
    img = np.zeros((n*w, n*w))
    for i, y in enumerate(np.linspace(*r1, n)):
        for j, x in enumerate(np.linspace(*r0, n)):
            z = torch.Tensor([[x, y]]).to(device)
            x_ = vae.decoder(z)
            x_ = x_.reshape(28, 28).to('cpu').detach().numpy()
            img[(n-1-i)*w:(n-1-i+1)*w, j*w:(j+1)*w] = x_
    plt.figure(figsize=(10,10))
    plt.imshow(img, extent=[*r0, *r1])
    
    
plot_reconstructed(vae)
```
