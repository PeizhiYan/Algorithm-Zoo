### The Neural Network Architecture (for binary classifiation):

---

$x_1$, $x_2$ are input features; $w_1$, $w_2$ are weights; $\sigma()$ is the sigmoid activation function. No bias.

---

$x_1$  $\longrightarrow$ $x_1 w_1$ 

$~~~~~~~~~~~~~~~~~~~~~~~~ \longrightarrow (x_1 w_1 + x_2 w_2) \longrightarrow \sigma(x_1 w_1 + x_2 w_2)$ 

$x_2$  $\longrightarrow$ $x_2 w_2$



```python
import numpy as np
np.random.seed(seed=369)
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
sns.set_style('whitegrid')
```

```python
""" synthesis some training data """

data = make_moons(n_samples=1000, shuffle=True, noise=0.05, random_state=None) # Gaussian noise std-dev = 0.05
X = data[0]
Y = np.reshape(data[1], [len(data[1]), 1])

# display first 5 samples
i = 0
for x,y in zip(X,Y):
    print('y =', y, '\t x =', x)
    i+=1
    if i>5:
        break

# Visualize 
color_book = np.array(['blue', 'red']) # blue for class 0, red for class 1

plt.figure(figsize=(8, 5))
plt.scatter(0,0,c='blue',label='Class 0',s=5) # legend for blue dots
plt.scatter(0,0,c='red',label='Class 1',s=5)  # legend for red dots
plt.scatter(0,0,c='white',s=5) # to "erase" the legend dummy dot
plt.scatter(X[:,0], X[:,1], c=color_book[np.reshape(Y, [Y.shape[0]])], s=1)
plt.legend()

plt.xlabel('Feature $x_1$', fontsize=22)
plt.ylabel('Feature $x_2$', fontsize=22)

plt.title('Training Samples Visualization', fontsize=18)
plt.legend(loc=2)
plt.show()
```

```python
def sigmoid(_Z):
    # _Z: [N, 1]
    # -----------------
    # sigmoid(z) = 1 / (1 + e^(-z))
    return 1 / (1 + np.e**(_Z))

def forward_propagation(_X, _W, activation=sigmoid):
    # _X: [N, D]
    # _W: [D, 1]
    # N is the number of samples
    # D is the input feature dimension
    # ---------------------------------
    # output = activation(XW)
    # sigmoid activation.
    return activation(_X @ _W)

def loss(_Y_hat, _Y):
    # binary cross-entropy loss
    return - np.mean(np.multiply(_Y, np.log(_Y_hat)) + np.multiply(1-_Y, np.log(1-_Y_hat)))

```

```python
""" 
 Plot the loss vs the different weight w1 and w2
 Note that, we only have two weights, so it is easy to plot the loss space as a 2D height-map
 If we have multiple weights (>3), it is impossible to visualize them together directly
"""

ws_1 = np.linspace(-5, 5, num=20)  # get 20 potential weight values (assume -5 <= w <= 5)
ws_2 = np.linspace(-5, 5, num=20)  # get 20 potential weight values (assume -5 <= w <= 5)
ws_x, ws_y = np.meshgrid(ws_1, ws_2) # generate grid
loss_ws = np.zeros((20,20))
for i in range(20):
    for j in range(20):
        loss_ws[i,j] = loss(forward_propagation(X, np.reshape(np.array([ws_1[i], ws_2[j]]), [2,1])), Y)

# Plot the loss space
plt.figure(figsize=(8, 5))
plt.contourf(ws_x, ws_y, loss_ws, 20)
plt.xlabel('$w_1$', fontsize=22)
plt.ylabel('$w_2$', fontsize=22)
plt.title('Loss Space', fontsize=22)
plt.colorbar()
plt.show()
```

**Gradient Descent:**   $ w^{(t+1)} = w^{(t)} - \mu \Delta w^{(t)} $ ($\mu$ is learning rate)

**Gradient:**   $ \Delta w = \cfrac{\partial L}{\partial w} $

**Our Neural Network $f(x)$:**   $ \hat{y} = f(x) = \sigma(z) = \sigma(wx) $

**Binary Cross-Entropy Loss (on $N$ samples):**   $ L = - \cfrac{1}{N}\sum \big[ ylog(\hat{y}) + (1 - y)log(1 -\hat{y}) \big] $

**Gradient (chain-rule):**   $ \cfrac{\partial L}{\partial w} = \cfrac{\partial L}{\partial \hat{y}} \cdot \cfrac{\partial \hat{y}}{\partial z} \cdot \cfrac{\partial z}{\partial w} = \cfrac{\hat{y} - y}{\hat{y}(1 - \hat{y})} \cdot \hat{y}(1 - \hat{y}) \cdot x = x \cdot (\hat{y} - y)$


```python
def compute_gradient(_W, _X, _Y):
    """ compute the gradient for weights in _W """
    return np.reshape(X.T @ (forward_propagation(_X, _W) - Y), [_W.shape[0],_W.shape[1]])


def compute_weight(_W, _X, _Y, learning_rate):
    """ calculate the updated weights in _W """
    return _W + learning_rate * compute_gradient(_W, _X, _Y)
```

```python
def dense_layer_weights(n_in, n_out, _min, _max):
    # generate random weights
    return (_max - _min) * np.reshape(np.random.rand(n_in*n_out), [n_in,n_out]) + _min
```

```python
EPOCHS = 15 # total epochs
LR = 1e-3 # learning rate

W = dense_layer_weights(n_in=2, n_out=1, _min=-1, _max=1) # randomly initialize the weights (-1.0~1.0)


""" optimize the weight (w) """
hist = [(W[0], W[1], loss(forward_propagation(X, W), Y))] # track weight and loss values
for i in range(EPOCHS):
    W = compute_weight(W, X, Y, LR) # update step
    l = loss(forward_propagation(X, W), Y)
    print('epoch ', i, ' loss ', l)
    hist.append((W[0][0], W[1][0], l))
    
plt.plot([i for i in range(len(hist))], [h[2] for h in hist], label='Train Loss')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Loss')
```

```python
""" Visualize Gradient Descent """
ws_1 = np.linspace(-2, 2, num=50)  # get 20 potential weight values 
ws_2 = np.linspace(-2, 2, num=50)  # get 20 potential weight values 
ws_x, ws_y = np.meshgrid(ws_1, ws_2) # generate grid
loss_ws = np.zeros((50,50))
for i in range(50):
    for j in range(50):
        loss_ws[i,j] = loss(forward_propagation(X, np.reshape(np.array([ws_1[j], ws_2[i]]), [2,1])), Y)

# Plot the loss space
plt.figure(figsize=(20, 20))
plt.contourf(ws_x, ws_y, loss_ws, 50)
plt.xlabel('$w_1$', fontsize=22)
plt.ylabel('$w_2$', fontsize=22)
plt.title('Loss Space', fontsize=22)
plt.colorbar()
#plt.show()

# Plot the updates
for i in range(0, len(hist)-1):
    w1_t1, w2_t1, l_t1 = hist[i]
    w1_t2, w2_t2, l_t2 = hist[i+1]
    plt.plot(w1_t1, w2_t1, 'bo')
    plt.plot([w1_t1, w1_t2], [w2_t1, w2_t2], 'g-', linewidth=4)
    plt.text(w1_t1, w2_t1, f't({i})', color='white', fontsize=18)

# Show figure
plt.xlabel('$w_1$', fontsize=22)
plt.ylabel('w_2', fontsize=22)
plt.title('Gradient descent updates plotted on loss space', fontsize=22)
plt.legend(loc=1)
plt.show()
```
