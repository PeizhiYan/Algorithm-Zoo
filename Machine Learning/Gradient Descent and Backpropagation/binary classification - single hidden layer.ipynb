```python
import numpy as np
np.random.seed(seed=369)
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
sns.set_style('whitegrid')
```

```python
""" synthesis some training data """

data = make_moons(n_samples=1000, shuffle=True, noise=0.05, random_state=None) # Gaussian noise std-dev = 0.05
X = data[0]
Y = np.reshape(data[1], [len(data[1]), 1])

# display first 5 samples
i = 0
for x,y in zip(X,Y):
    print('y =', y, '\t x =', x)
    i+=1
    if i>5:
        break

# Visualize 
color_book = np.array(['blue', 'red']) # blue for class 0, red for class 1

plt.figure(figsize=(8, 5))
plt.scatter(0,0,c='blue',label='Class 0',s=5) # legend for blue dots
plt.scatter(0,0,c='red',label='Class 1',s=5)  # legend for red dots
plt.scatter(0,0,c='white',s=5) # to "erase" the legend dummy dot
plt.scatter(X[:,0], X[:,1], c=color_book[np.reshape(Y, [Y.shape[0]])], s=1)
plt.legend()

plt.xlabel('Feature $x_1$', fontsize=22)
plt.ylabel('Feature $x_2$', fontsize=22)

plt.title('Training Samples Visualization', fontsize=18)
plt.legend(loc=2)
plt.show()
```

```python
def sigmoid(_Z):
    # _Z: [N, 1]
    # -----------------
    # sigmoid(z) = 1 / (1 + e^(-z))
    return 1 / (1 + np.e**(_Z))

def relu(_Z):
    # _Z: [N, 1]
    # -----------------
    # ReLU(z) = max(0, _Z)
    return _Z * (_Z > 0)

def linear(_Z):
    # linear activation
    return _Z

def forward(_X, _W, activation=sigmoid):
    # _X: [N, D]
    # _W: [D, 1]
    # N is the number of samples
    # D is the input feature dimension
    # ---------------------------------
    # output = activation(XW)
    return activation(_X @ _W)

def loss(_Y_hat, _Y):
    # binary cross-entropy loss
    return - np.mean(np.multiply(_Y, np.log(_Y_hat)) + np.multiply(1-_Y, np.log(1-_Y_hat)))

```

reference: https://sudeepraja.github.io/Neural/

**Gradient Descent:**   $ w^{(t+1)} = w^{(t)} - \mu \Delta w^{(t)} $ ($\mu$ is learning rate)

**Gradient:**   $ \Delta w = \cfrac{\partial L}{\partial w} $

**Our Neural Network $f(x)$:**   

$ \hat{y} = f(x) = \sigma(z_o) = \sigma(z_h W_o)$

$ z_h = ReLU(o_h) = ReLU(x W_h)$

**Binary Cross-Entropy Loss (on $N$ samples):**   $ L(\hat{y},y) = - \cfrac{1}{N}\sum \big[ ylog(\hat{y}) + (1 - y)log(1 -\hat{y}) \big] $

**Gradient (chain-rule):**   

$ \cfrac{\partial L}{\partial W_o} = \cfrac{\partial L}{\partial \hat{y}} \cdot \cfrac{\partial \hat{y}}{\partial z_o}  \cdot \cfrac{\partial z_o}{\partial W_o} = \cfrac{\hat{y} - y}{\hat{y}(1 - \hat{y})} \cdot \hat{y}(1 - \hat{y}) \cdot z_h = (\hat{y} - y) \cdot z_h $

$ \cfrac{\partial L}{\partial W_h} = \cfrac{\partial L}{\partial \hat{y}} \cdot \cfrac{\partial \hat{y}}{\partial z_o} \cdot \cfrac{\partial z_o}{\partial z_h} \cdot \cfrac{\partial z_h}{\partial o_h} \cdot \cfrac{\partial o_h}{\partial W_h} = \cfrac{\hat{y} - y}{\hat{y}(1 - \hat{y})} \cdot \hat{y}(1 - \hat{y}) \cdot W_o \cdot ReLU'(o_h) \cdot x = (\hat{y} - y) \cdot W_o \cdot ReLU'(o_h) \cdot x $

$ReLU'(o_h) = 1$     when  $o_h > 0$

$ReLU'(o_h) = 0$     when  $o_h \leq 0$


```python
def drelu(_X):
    """ derivative of ReLU """
    # drelu(x) = 1  if x > 0
    # drelu(x) = 0  if x <= 0
    return 1. * (_X > 0)

def compute_gradient(_Wh, _Wo, _X, _Y):
    """ compute gradient for both hidden and output layer weights """
    _oh = forward(_X, _Wh, linear) # [n x d]
    _Zh = relu(_oh) # [n x d]
    _Y_hat = forward(_Zh, _Wo, sigmoid) # [n x 1]
    
    # compute the gradient for output layer weights in _Wo
    delta_Wo = np.reshape(_Zh.T @ (_Y_hat - Y), [_Wo.shape[0],_Wo.shape[1]]) # [d x 1]
    
    # compute the gradient for hidden layer weights in _Wh
    _XT = _X.T  # [2 x n]
    _YY = _Y_hat - Y # [n x 1]
    _YYT = _YY.T # [1 x n]
    delta_Wh = _XT @ (drelu(_oh).T * (_Wo @ _YYT)).T # [2 x d]
    
    return delta_Wh, delta_Wo

def compute_weight(_Wh, _Wo, _X, _Y, learning_rate):
    """ calculate the updated weights """
    delta_Wh, delta_Wo = compute_gradient(_Wh, _Wo, _X, _Y)
    _Wh = _Wh + learning_rate * delta_Wh
    _Wo = _Wo + learning_rate * delta_Wo    
    return _Wh, _Wo
```

```python
def dense_layer_weights(n_in, n_out, _min, _max):
    """ generate random weights """
    return (_max - _min) * np.reshape(np.random.rand(n_in*n_out), [n_in,n_out]) + _min
```

```python
def nn_predict(_Wh, _Wo, _X, return_score=True):
    """ forward pass through the entire network """
    scores = forward(forward(_X, _Wh, relu), _Wo, sigmoid)
    if return_score:
        return scores
    else:
        return 1 * (scores > 0.5)
```

```python
EPOCHS = 200 # total epochs
LR = 1e-3 # learning rate

# hidden layer weights
Wh = dense_layer_weights(n_in=2, n_out=6, _min=-1, _max=1) # randomly initialize the weights (-1.0~1.0)

# output layer weights
Wo = dense_layer_weights(n_in=6, n_out=1, _min=-1, _max=1) # randomly initialize the weights (-1.0~1.0)


""" optimize the weight (w) """
hist = [loss(nn_predict(Wh, Wo, X), Y)] # track loss values
for i in range(EPOCHS):
    Wh, Wo = compute_weight(Wh, Wo, X, Y, LR) # update step
    l = loss(nn_predict(Wh, Wo, X), Y)
    if i % 10 == 0:
        print('epoch ', i, ' loss ', l)
    hist.append(l)
    
plt.plot([i for i in range(len(hist))], [h for h in hist], label='Train Loss')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Loss')
```

```python
def accuracy(y_pred, y_true):
    return (y_pred == y_true).mean()

y_pred = nn_predict(Wh, Wo, X, return_score=False)

print("Train accuracy: ", accuracy(y_pred, Y))
```

```python
""" Visualize Decision Boundary """

x_1 = np.linspace(-2, 2, num=400)
x_2 = np.linspace(-2, 2, num=400)
X_prob = []
for i in range(len(x_1)):
    for j in range(len(x_2)):
        X_prob.append([x_1[j], x_2[i]])
X_prob = np.array(X_prob)
Y_prob = nn_predict(Wh, Wo, X_prob, return_score=True)
decision_map = np.zeros([len(x_1),len(x_2)])
for i in range(len(x_1)):
    for j in range(len(x_2)):
        decision_map[i,j] = Y_prob[i*len(x_1)+j]

# generate 2 2d grids for the x & y bounds
x, y = np.meshgrid(x_2, x_1)
z = -(decision_map-0.5)*2
# x and y are bounds, so z should be the value *inside* those bounds.
# Therefore, remove the last value from the z array.
z = z[:-1, :-1]
z_min, z_max = -np.abs(z).max(), np.abs(z).max()

fig, ax = plt.subplots()
c = ax.pcolormesh(x, y, z, cmap='RdBu', vmin=z_min, vmax=z_max)
# set the limits of the plot to the limits of the data
ax.axis([x.min(), x.max(), y.min(), y.max()])
Y_PRED = nn_predict(Wh, Wo, X, return_score=False)
plt.scatter(0,0,c='blue',label='Class 0',s=5) # legend for blue dots
plt.scatter(0,0,c='red',label='Class 1',s=5)  # legend for red dots
plt.scatter(X[:,0], X[:,1], c=color_book[np.reshape(Y_PRED, [Y_PRED.shape[0]])], s=1)
plt.legend()
plt.xlabel('Feature $x_1$', fontsize=22)
plt.ylabel('Feature $x_2$', fontsize=22)
plt.title('Decision Boundary and Predicted Labels', fontsize=22)
plt.show()

plt.scatter(0,0,c='blue',label='Class 0',s=5) # legend for blue dots
plt.scatter(0,0,c='red',label='Class 1',s=5)  # legend for red dots
plt.scatter(X[:,0], X[:,1], c=color_book[np.reshape(Y, [Y.shape[0]])], s=1)
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.legend()
plt.xlabel('Feature $x_1$', fontsize=22)
plt.ylabel('Feature $x_2$', fontsize=22)
plt.title('True Labels', fontsize=22)
plt.show()
```
