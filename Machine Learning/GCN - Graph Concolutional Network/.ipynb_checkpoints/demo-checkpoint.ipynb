# GCN via Tensorflow v1: based on Kipf's paper

```python
import tensorflow as tf
import numpy as np
import scipy.sparse as sp
from scipy.sparse.linalg.eigen.arpack import eigsh

##############################################
# numpy matrix processing utility functions ##
##############################################
def sparse_to_tuple(sparse_mx):
    """ Convert sparse matrix to tuple representation. """
    def to_tuple(mx):
        if not sp.isspmatrix_coo(mx):
            mx = mx.tocoo()
        coords = np.vstack((mx.row, mx.col)).transpose()
        values = mx.data
        shape = mx.shape
        return coords, values, shape
    if isinstance(sparse_mx, list):
        for i in range(len(sparse_mx)):
            sparse_mx[i] = to_tuple(sparse_mx[i])
    else:
        sparse_mx = to_tuple(sparse_mx)
    return sparse_mx
def normalize_adj(adj):
    """ 
        Symmetrically normalize adjacency matrix. 
        Equation:
            S = D^(-1/2) * A * D^(-1/2)
    """
    adj = sp.coo_matrix(adj)
    rowsum = np.array(adj.sum(1))
    d_inv_sqrt = np.power(rowsum, -0.5).flatten()
    d_inv_sqrt[np.isinf(d_inv_sqrt)] = 0.
    d_mat_inv_sqrt = sp.diags(d_inv_sqrt)
    return adj.dot(d_mat_inv_sqrt).transpose().dot(d_mat_inv_sqrt).tocoo()
def preprocess_adj(adj):
    """ 
        Preprocessing of adjacency matrix for simple GCN model and conversion to tuple representation. 
        return a sparse support matrix
    """
    adj_normalized = normalize_adj(adj + sp.eye(adj.shape[0]))
    return sparse_to_tuple(adj_normalized)

##############################################
# create Tensor variables utility functions ##
##############################################
def uniform(shape, scale=0.05, name=None):
    """Uniform init."""
    initial = tf.random_uniform(shape, minval=-scale, maxval=scale, dtype=tf.float32)
    return tf.Variable(initial, name=name)
def glorot(shape, name=None):
    """Glorot & Bengio (AISTATS 2010) init."""
    init_range = np.sqrt(6.0/(shape[0]+shape[1]))
    initial = tf.random_uniform(shape, minval=-init_range, maxval=init_range, dtype=tf.float32)
    return tf.Variable(initial, name=name)
def zeros(shape, name=None):
    """All zeros."""
    initial = tf.zeros(shape, dtype=tf.float32)
    return tf.Variable(initial, name=name)
def ones(shape, name=None):
    """All ones."""
    initial = tf.ones(shape, dtype=tf.float32)
    return tf.Variable(initial, name=name)

#########################################
# Tensor computation utility functions ##
#########################################
def dot(x, y, sparse=False):
    """ Wrapper for tf.matmul """
    if sparse:
        return tf.sparse_tensor_dense_matmul(x, y) # x: sparse; y: dense
    else:
        return tf.matmul(x, y)
    
####################
# layer functions ##
####################
def dense_layer(x, in_dim, out_dim, activation=tf.nn.relu, bias=False):
    """ create a dense layer """
    w = glorot([in_dim, out_dim])
    h = dot(x, w)
    if bias:
        b = zeros([out_dim])
        h = h + b
    return activation(h)
def graph_conv_layer(x, support, in_dim, out_dim, activation=tf.nn.relu, sparse_input=False, bias=False):
    """ create a graph convolution layer """
    w = glorot([in_dim, out_dim])
    pre_sup = dot(x, w, sparse=sparse_input)
    h = dot(support, pre_sup, sparse=True) # NOTE: support is sparse; pre_sup is dense
    if bias:
        b = zeros([out_dim])
        h = h + b
    return activation(h)

```

### Test the encoding of GCN w/o training

```python
tf.reset_default_graph()

N = len(A) # number of nodes
C = len(A) # number of features

support_ph = tf.sparse_placeholder(tf.float32) # support matrix placeholder. H = S * X * W
feature_ph = tf.placeholder(tf.float32, shape=[N,C]) # feature matrix placeholder

h0 = feature_ph

h1 = graph_conv_layer(x=h0, support=support_ph, in_dim=C, out_dim=32, activation=tf.nn.relu)
h2 = graph_conv_layer(x=h1, support=support_ph, in_dim=32, out_dim=32, activation=tf.nn.relu)
h3 = graph_conv_layer(x=h2, support=support_ph, in_dim=32, out_dim=2, activation=tf.nn.sigmoid)

out = h3

""" prepare some experimental data (from Karate Club Graph) """
#fea = np.array([[1.0] for i in range(len(A))]) # the fake feature matrix
fea = np.eye(len(A))
sup = preprocess_adj(A) # the sparse support matrix

sess = tf.Session()
sess.run(tf.global_variables_initializer())
gcn_output = sess.run(out, feed_dict={support_ph: sup, feature_ph: fea})

""" visualize the 2D output """
x = [gcn_output[i,0] for i in range(len(order))]
y = [gcn_output[i,1] for i in range(len(order))]
plt.scatter(x=x, y=y, color=color)
```

### Test the node-level classfication ability of GCN w/ training

```python
plt.figure(figsize=(15,5))

tf.reset_default_graph()

N = len(A) # number of nodes
C = len(A) # number of features

support_ph = tf.sparse_placeholder(tf.float32) # support matrix placeholder. H = S * X * W
feature_ph = tf.placeholder(tf.float32, shape=[N,C]) # feature matrix placeholder
label_ph = tf.placeholder(tf.int64, shape=[N]) # the labels placeholder

label_onehot = tf.one_hot(label_ph, depth=2) # convert indices (labels) to one-hot labels 

h0 = feature_ph

h1 = graph_conv_layer(x=h0, support=support_ph, in_dim=C, out_dim=32, activation=tf.nn.relu)
h2 = graph_conv_layer(x=h1, support=support_ph, in_dim=32, out_dim=32, activation=tf.nn.relu)
h3 = graph_conv_layer(x=h2, support=support_ph, in_dim=32, out_dim=2, activation=tf.nn.sigmoid)

out = h3

loss = tf.nn.softmax_cross_entropy_with_logits(labels=label_onehot, logits=out) # cross-entropy loss
loss = tf.reduce_mean(loss)

opt =  tf.train.AdamOptimizer(learning_rate=1e-3).minimize(loss)


""" prepare some experimental data (from Karate Club Graph) """
#fea = np.array([[1.0] for i in range(len(A))]) # the fake feature matrix
fea = np.eye(len(A))
sup = preprocess_adj(A) # the sparse support matrix

sess = tf.Session()
sess.run(tf.global_variables_initializer())

""" visualize the 2D output """
gcn_output = sess.run(out, feed_dict={support_ph: sup, feature_ph: fea}) # this output is without training
x = [gcn_output[i,0] for i in range(len(order))]
y = [gcn_output[i,1] for i in range(len(order))]
plt.subplot(1,3,1)
plt.scatter(x=x, y=y, color=color)
plt.title("Visualization w/o Train")

""" training """
loss_hist = [] # loss history
for epoch in range(1000):
    _, loss_value = sess.run([opt,loss], feed_dict={support_ph: sup, feature_ph: fea, label_ph: labels})
    loss_hist.append(loss_value)
print("final loss: ", loss_value)
plt.subplot(1,3,2)
plt.plot(loss_hist)
plt.title("Loss")

""" visualize the 2D output """
gcn_output = sess.run(out, feed_dict={support_ph: sup, feature_ph: fea}) # this output is without training
x = [gcn_output[i,0] for i in range(len(order))]
y = [gcn_output[i,1] for i in range(len(order))]
plt.subplot(1,3,3)
plt.scatter(x=x, y=y, color=color)
plt.title("Visualization w/ Train")
```
