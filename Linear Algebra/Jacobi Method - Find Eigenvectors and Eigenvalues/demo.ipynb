```python
import matplotlib.pyplot as plt
import cv2
import h5py
import numpy as np
import math
```

```python
def jacobi_eig(x, tol=1e-9):
    """
    Jacobi method : find eigenvectors and eigenvalues of a square matrix
    WARNING: This implementation is not efficient!
        x - square matrix
    Returns:
        eigenvalues
        eigenvectors
    """
    def off_diag_max(x):
        """ 
        Find large off-diagonal element
            x - square matrix
        returns:
            dmax - maximum value
               i - row index
               j - column index
        """        
        n = x.shape[0]
        omax = .0
        for i in range(n-1):
            for j in range(i+1, n):
                if abs(x[i,j]) >= omax:
                    omax = abs(x[i,j])
                    k = i
                    l = j
        return omax, k, l

    def rotate(x, p, k, l):
        """
        Rotate to make x[k,l] == 0
        """
        n = x.shape[0]
        diff = x[l,l] - x[k,k]
        if abs(x[k,l]) < abs(diff)*1e-36:
            t = x[k,l]/diff
        else:
            phi = diff/(2.0*x[k,l])
            t = 1.0/(abs(phi) + math.sqrt(phi**2 + 1.0))
            if phi < .0:
                t = -t
        c = 1.0/math.sqrt(t**2 + 1.0)
        s = t*c
        tau = s/(1.0 + c)
        tmp = x[k,l]
        x[k,l] = .0
        x[k,k] = x[k,k] - t*tmp
        x[l,l] = x[l,l] + t*tmp
        for i in range(k):
            tmp = x[i,k]
            x[i,k] = tmp - s*(x[i,l] + tau*tmp)
            x[i,l] = x[i,l] + s*(tmp - tau*x[i,l])
        for i in range(k+1,l):
            tmp = x[k,i]
            x[k,i] = tmp - s*(x[i,l] + tau*x[k,i])
            x[i,l] = x[i,l] + s*(tmp - tau*x[i,l])
        for i in range(l+1,n):
            tmp = x[k,i]
            x[k,i] = tmp - s*(x[l,i] + tau*tmp)
            x[l,i] = x[l,i] + s*(tmp - tau*x[l,i])
        for i in range(n):
            tmp = p[i,k]
            p[i,k] = tmp - s*(p[i,l] + tau*p[i,k])
            p[i,l] = p[i,l] + s*(tmp - tau*p[i,l])
    
    n = x.shape[0]
    rotMax = 5*(n**2) # set limit on the number of rotations
    p = 1.0*np.eye(n) # initialize transformation matrix
    for i in range(rotMax):
        omax, k, l = off_diag_max(x)
        if omax < tol:
            return np.diagonal(x), p # Eigenvalues, Eigenvectors
        rotate(x, p, k, l)
    assert Exception('did not converge')
    print('did not converge')
    
    
    
```

```python
# Load USPS handwritten digits
with h5py.File("./dataset/usps.h5", 'r') as hf:
    train = hf.get('train')
    train_x = train.get('data')[:1000]
print(train_x.shape)

images = np.zeros([train_x.shape[0], 8*8])
for i in range(len(train_x)):
    im = cv2.resize(np.reshape(train_x[i], [16,16]), (8,8))
    images[i,:] = np.reshape(im, [-1, 8*8])[:]
    
# Display some samples
plt.figure(figsize=(10,2))
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(np.reshape(images[i],[8,8]), cmap='gray')
    plt.axis("off")
```

```python
# Calculate the covariance matrix of the dataset
x = np.copy(images)
mu = np.mean(x, axis=0) # Compute mean for each dimension
x = x - mu # Subtract off the mean for each dimension
cov = 1 / (x.shape[0]-1) * np.matmul(np.transpose(x), x)

plt.imshow(cov, cmap='gray')
```

```python
# Use Jacobi method to find eigenvectors and eigenvalues
v, pc = jacobi_eig(cov)
```

```python
# Sort the eigenvectors in terms of eigenvalues
s = np.argsort(v)[::-1]
pc[:,:] = pc[s,:]
pc[:,:] = pc[:,s]
```

```python
# Show the 64 principle components (looks kinda like the 2D discreate cosine transform basis :)
plt.figure(figsize=(8,8))
for i in range(8):
    for j in range(8):
        plt.subplot(8,8,i*8+j+1)
        plt.imshow(np.reshape(pc[:,i*8+j],[8,8]), cmap='gray')
        plt.axis("off")
```

```python
# Use Numpy's eig() function
(v, pc) = np.linalg.eig(cov)

# Show the 64 principle components (looks kinda like the 2D discreate cosine transform basis :)
plt.figure(figsize=(8,8))
for i in range(8):
    for j in range(8):
        plt.subplot(8,8,i*8+j+1)
        plt.imshow(np.reshape(pc[:,i*8+j],[8,8]), cmap='gray')
        plt.axis("off")
```
