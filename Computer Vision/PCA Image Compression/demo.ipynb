```python
import numpy as np
import matplotlib.pyplot as plt
import pca_img_compress as PIC
import cv2
```

```python
# Load and display the original image   
# IMAGE COPY RIGHT: Peizhi Yan
img = PIC.load_bmp('./data/img.bmp')
print(img.shape)
plt.figure(figsize=(10,10))
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# Compress with 16 principle components
PIC.compress(img=img, save_path='./data/img_16pc', K=16)
img_16pc_reconstruct = PIC.reconstruct(load_path='./data/img_16pc')
cv2.imwrite('./data/img_16pc.bmp', img_16pc_reconstruct)

plt.figure(figsize=(10,10))
plt.imshow(img_16pc_reconstruct, cmap='gray')
plt.title('PC = 16')
plt.show()
```
