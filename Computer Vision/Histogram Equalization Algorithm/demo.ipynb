```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load and display an image
img = cv2.imread('./images/face4.bmp', 0)
plt.imshow(img, cmap='gray')
plt.colorbar()
print(img.shape)
```

```python
HEIGHT = img.shape[0] # get the height of the image
WIDTH = img.shape[1] # get the width of the image
```

```python
img_ = np.copy(img) # create a copy of the original image
```

```python
# Calculate the histogram of the original image (assume it is gray scale image)

hist = np.zeros([256]) # pixel intensity should be integers within 0 ~ 255

for y in range(HEIGHT):
    for x in range(WIDTH):
        pixel = img[y][x] # get the pixel intensity
        hist[pixel] += 1 # count

# Display the original histogram
plt.bar(x=[i for i in range(256)], height=hist)
plt.xlabel("Pixel Intensity")
plt.ylabel("Count")
```

```python
# Compute the cumulative histogram
cum_hist = np.copy(hist)
for idx in range(1,256):
    cum_hist[idx] = hist[idx] + cum_hist[idx-1]

# Display the cumulative histogram
plt.plot(cum_hist)
plt.xlabel("Pixel Intensity")
plt.ylabel("Cumulative Count")
```

```python
# Divide the cumulative histogram by the number of pixels
cum_hist /= (WIDTH*HEIGHT) # so the values will be in 0~1

# Display the normalized cumulative histogram
plt.plot(cum_hist)
plt.xlabel("Pixel Intensity")
plt.ylabel("Count")
```

```python
# Map new pixel intensity to the result image
for y in range(HEIGHT):
    for x in range(WIDTH):
        pixel = int(img[y][x])
        img_[y][x] = 255*(cum_hist[pixel])
```

```python
# Display the results
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.colorbar()
plt.title("Original Image")
plt.subplot(2,2,3)
_ = plt.hist(np.reshape(img, [-1]), bins=256)
plt.title("Original Histogram")

plt.subplot(2,2,2)
plt.imshow(img_, cmap='gray')
plt.colorbar()
plt.title("Result")
plt.subplot(2,2,4)
_ = plt.hist(np.reshape(img_, [-1]), bins=256)
plt.title("Result Histogram")
```
