```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
"""
Algorithm Description:
    In this demo, I design a very simple method for hiding an image
    to a cover image.
"""

def hide(secret_image_path, cover_image_path):
    # Note: Better to have the secret image and cover image have the same shape!
    #       The container image should never be converted to integer type! 
    
    # load secret image 
    secret_img = cv2.imread(secret_image_path)
    secret_img = cv2.cvtColor(secret_img, cv2.COLOR_BGR2RGB)
    
    # load cover image
    cover_img = cv2.imread(cover_image_path)
    cover_img = cv2.cvtColor(cover_img, cv2.COLOR_BGR2RGB)
    
    # resize secret image to the same size of cover image
    secret_img = cv2.resize(secret_img, (cover_img.shape[1],cover_img.shape[0]), interpolation = cv2.INTER_AREA)
    
    # if cover image value is 255., subtract by 1 to make some space for secret image
    cover_img[cover_img == 255] = 255 - 1
    
    # convert both images to floating point arraies
    secret_img = np.array(secret_img, dtype=float)
    cover_img = np.array(cover_img, dtype=float)
    
    # scale the value of scret image from 0.~255. to 0.~1.0
    secret_img = secret_img / 255.
    
    # add rescaled secret image to cover iamge to get container image
    container_image = secret_img + cover_img
    
    # normalize the value of container iamge to 0.~1.0
    container_image = container_image / 255.
    
    return container_image
    
    
def reveal(container_image):
    # rescale container image to 0.~255.
    container_image = container_image * 255.
    
    # only keep the fractional part
    container_image_int_part = np.array(container_image, dtype=int)
    recovered_secret_image = container_image - container_image_int_part
    
    # reshace secret image to 0.~255.
    recovered_secret_image = recovered_secret_image * 255.
    recovered_secret_image = np.array(recovered_secret_image, int)
    
    return recovered_secret_image
    
```

```python
""" Display Original Images """
fox_img = cv2.imread('./data/fox.PNG')
fox_img = cv2.cvtColor(fox_img, cv2.COLOR_BGR2RGB)
tiger_img = cv2.imread('./data/tiger.PNG')
tiger_img = cv2.cvtColor(tiger_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(fox_img)
plt.subplot(1,2,2)
plt.imshow(tiger_img)
```

```python
""" Hide Fox image into Tiger image """
container_image = hide(secret_image_path = './data/fox.PNG', cover_image_path = './data/tiger.PNG')

recovered_secret_image = reveal(container_image)

plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(container_image)
plt.title("Container Image")
plt.subplot(1,2,2)
plt.imshow(recovered_secret_image)
plt.title("Recovered Secret Image")
```

```python
""" Hide Tiger image into Fox image """
container_image = hide(secret_image_path = './data/tiger.PNG', cover_image_path = './data/fox.PNG')

recovered_secret_image = reveal(container_image)

plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(container_image)
plt.title("Container Image")
plt.subplot(1,2,2)
plt.imshow(recovered_secret_image)
plt.title("Recovered Secret Image")
```
